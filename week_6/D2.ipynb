```python
# take any three feature and build regression  
```

```python
x = df[['cyl','disp','hp']]
y = df[['mpg']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
print('r2 score:',lr.score(x,y))
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
fs= SequentialFeatureSelector (LinearRegression(),
                              direction ='backward')
```

```python
x = df[['cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear',
       'carb']]
y = df[['mpg']]   # option : x = df.drop('mpg',axis=1)
```

```python
fs.fit(x,y)
```

```python
fs.get_support()
```

```python
x = x.iloc[:,fs.get_support()]
```

```python
x
```

```python
lr.fit(x,y)
```

```python
lr.score(x,y)
```
