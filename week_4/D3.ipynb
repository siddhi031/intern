 ## Use agg() to apply multiple aggregation functions (mean, min, max) on a grouped DataFrame

```python
import pandas as pd

data = {
    'name': ['Aarav', 'Priya', 'Rohan', 'Ananya', 'Vikram'],
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Mumbai', 'Chennai'],
    'marks': [85, 90, 78, 88, 92]
}

df5 = pd.DataFrame(data)
df5
```

```python
grouped = df5.groupby('city')['marks']
mean_marks = grouped.mean()
print("Mean Marks:")
mean_marks
```

```python
min_marks = grouped.min()
print("Minimum Marks:")
min_marks
```

```python

max_marks = grouped.max()
print("Maximum Marks:")
max_marks
```

## Merge two DataFrames on a common column (e.g., student ID).

```python
import pandas as pd

students = pd.DataFrame({
    'student_id': [1, 2, 3],
    'name': ['Aarav', 'Priya', 'Rohan'],
    'city': ['Delhi', 'Mumbai', 'Chennai']
})
students

```

```python
marks = pd.DataFrame({
    'student_id': [1, 2, 3],
    'marks': [85, 90, 78]
})
marks

```

```python
merged_df = pd.merge(students, marks, on='student_id')

merged_df
```
